```python
import time
import warnings
warnings.filterwarnings("ignore")
start = time.process_time()

prom_len_2oM = np.zeros([df_directorio_1oB.RBD.shape[0]])
prom_mat_2oM = np.zeros([df_directorio_1oB.RBD.shape[0]])
pend_len_2oM = np.zeros([df_directorio_1oB.RBD.shape[0]])
pend_mat_2oM = np.zeros([df_directorio_1oB.RBD.shape[0]])

prom_len_4oB = np.zeros([df_directorio_1oB.RBD.shape[0]])
prom_mat_4oB = np.zeros([df_directorio_1oB.RBD.shape[0]])
pend_len_4oB = np.zeros([df_directorio_1oB.RBD.shape[0]])
pend_mat_4oB = np.zeros([df_directorio_1oB.RBD.shape[0]])

for idx, rbd in enumerate(df_directorio_1oB.RBD):

    df_simce_2oM_n = df_simce_2oM[df_simce_2oM['RBD']==rbd].copy()
    df_simce_4oB_n = df_simce_4oB[df_simce_4oB['RBD']==rbd].copy()

    if df_simce_2oM_n.shape[0]>0:
        prom_len_2oM[idx] = np.mean(df_simce_2oM_n.PROM_LECT2M_RBD)
        prom_mat_2oM[idx] = np.mean(df_simce_2oM_n.PROM_MATE2M_RBD)

        # pendiente
        agno = np.unique(df_simce_2oM_n.AGNO)
        prom_vector_l = np.zeros(agno.shape[0])
        prom_vector_m = np.zeros(agno.shape[0])
        agno_vector = np.zeros(agno.shape[0])
        for a in range(agno.shape[0]):
            df_temp_new = df_simce_2oM_n[df_simce_2oM_n.AGNO==agno[a]].copy()
            prom_vector_l[a] = np.mean(df_temp_new.PROM_LECT2M_RBD)
            prom_vector_m[a] = np.mean(df_temp_new.PROM_MATE2M_RBD)
            agno_vector[a] = a+1
        x = agno_vector

        if (x.shape[0]*np.nansum(x**2.) - (np.nansum(x))**2.) == 0.:
          pend_len_2oM[idx] = 0.
          pend_mat_2oM[idx] = 0.
        else:
          y = prom_vector_l
          pend_len_2oM[idx] = (x.shape[0]*(np.nansum(x*y))-(np.nansum(x)*np.nansum(y)))/(x.shape[0]*np.nansum(x**2.) - (np.nansum(x))**2.)

          y = prom_vector_m
          pend_mat_2oM[idx] = (x.shape[0]*(np.nansum(x*y))-(np.nansum(x)*np.nansum(y)))/(x.shape[0]*np.nansum(x**2.) - (np.nansum(x))**2.)

    else:
        prom_len_2oM[idx] = np.nan
        prom_mat_2oM[idx] = np.nan
        pend_len_2oM[idx] = np.nan
        pend_mat_2oM[idx] = np.nan


    if df_simce_4oB_n.shape[0]>0:
        prom_len_4oB[idx] = np.mean(df_simce_4oB_n.PROM_LECT4B_RBD)
        prom_mat_4oB[idx] = np.mean(df_simce_4oB_n.PROM_MATE4B_RBD)

        # pendiente
        agno = np.unique(df_simce_4oB_n.AGNO)
        prom_vector_l = np.zeros(agno.shape[0])
        prom_vector_m = np.zeros(agno.shape[0])
        agno_vector = np.zeros(agno.shape[0])
        for a in range(agno.shape[0]):
            df_temp_new = df_simce_4oB_n[df_simce_4oB_n.AGNO==agno[a]].copy()
            prom_vector_l[a] = np.mean(df_temp_new.PROM_LECT4B_RBD)
            prom_vector_m[a] = np.mean(df_temp_new.PROM_MATE4B_RBD)
            agno_vector[a] = a+1
        x = agno_vector

        if (x.shape[0]*np.nansum(x**2.) - (np.nansum(x))**2.) == 0.:
          pend_len_4oB[idx] = 0.
          pend_mat_4oB[idx] = 0.
        else:
          y = prom_vector_l
          pend_len_4oB[idx] = (x.shape[0]*(np.nansum(x*y))-(np.nansum(x)*np.nansum(y)))/(x.shape[0]*np.nansum(x**2.) - (np.nansum(x))**2.)

          y = prom_vector_m
          pend_mat_4oB[idx] = (x.shape[0]*(np.nansum(x*y))-(np.nansum(x)*np.nansum(y)))/(x.shape[0]*np.nansum(x**2.) - (np.nansum(x))**2.)

    else:
        prom_len_4oB[idx] = np.nan
        prom_mat_4oB[idx] = np.nan
        pend_len_4oB[idx] = np.nan
        pend_mat_4oB[idx] = np.nan


```
