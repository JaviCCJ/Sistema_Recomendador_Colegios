```python
def extraccion_datos_df(df):
    ################################################################################
    promedio = np.zeros([12]) # calculado sobre todos los años
    asistencia = np.zeros([12]) # calculado sobre todos los años
    porcentaje_aprob = np.zeros([12]) # calculado sobre todos los años
    porcentaje_nodesercion = np.zeros([12]) # calculado sobre todos los años

    porcentaje_hombres = np.zeros([12]) # calculado solo para el 2020
    cantidad_alumnos = np.zeros([12]) # calculado solo para el 2020
    cantidad_cursos = np.zeros([12]) # calculado solo para el 2020
    cantidad_hombres = np.zeros([12]) # calculado solo para el 2020
    cantidad_mujeres = np.zeros([12]) # calculado solo para el 2020

    pendiente_promedio = np.zeros([12]) # calculado sobre todos los años

    ################################################################################
    id_ense = np.unique(df.COD_ENSE)

    if np.sum(id_ense==110)>0:
        # tengo que buscar cursos de basica, partiendo de primero basico, si no lo encuentro no se imparte.
        for n in range(8):
            df_temp = df[(df.COD_ENSE==110) & (df.COD_GRADO==n+1)]
            if df_temp.shape[0]>0:

                #################################################################################################################################
                # realizo los calculos
                # promedio de notas
                promedio[n] = np.mean(df_temp.PROM_GRAL)

                # promedio asistencia
                asistencia[n] = np.mean(df_temp.ASISTENCIA)

                # promedio de aprobados vs reprobados
                num = (np.sum(df_temp.SIT_FIN_R=='P') - np.sum(df_temp.SIT_FIN_R=='R'))
                deno = (np.sum(df_temp.SIT_FIN_R=='P') + np.sum(df_temp.SIT_FIN_R=='R'))
                if deno>0:
                    porcentaje_aprob[n] = (num/deno)*100
                else:
                    porcentaje_aprob[n] = np.nan

                # porcentaje de retencion
                num = (np.sum(df_temp.SIT_FIN_R=='P') + np.sum(df_temp.SIT_FIN_R=='R') - np.sum(df_temp.SIT_FIN_R=='Y') - np.sum(df_temp.SIT_FIN_R=='T'))
                deno = (np.sum(df_temp.SIT_FIN_R=='P') + np.sum(df_temp.SIT_FIN_R=='R') + np.sum(df_temp.SIT_FIN_R=='Y') + np.sum(df_temp.SIT_FIN_R=='T'))
                if deno>0:
                    porcentaje_nodesercion[n] = (num/deno)*100
                else:
                    porcentaje_nodesercion[n] = np.nan

                # pendiente
                agno = np.unique(df_temp.AGNO)
                prom_vector = np.zeros(agno.shape[0])
                agno_vector = np.zeros(agno.shape[0])
                for a in range(agno.shape[0]):
                    df_temp_new = df_temp[ df_temp.AGNO == agno[a] ].copy()
                    prom_vector[a] = np.mean(df_temp_new.PROM_GRAL)
                    agno_vector[a] = a+1
                x = agno_vector
                y = prom_vector

                pendiente_promedio[n] = (x.shape[0]*(np.sum(x*y))-(np.sum(x)*np.sum(y)))/(x.shape[0]*np.sum(x**2.) - (np.sum(x))**2.)

                #################################################################################################################################
                df_temp_new = df_temp[df_temp.AGNO==2020].copy()
                if df_temp_new.shape[0]>0:

                    # porcentaje de hombres con respecto a mujeres
                    porcentaje_hombres[n] = (np.sum(df_temp_new.GEN_ALU==1)/(np.sum(df_temp_new.GEN_ALU==1) + np.sum(df_temp_new.GEN_ALU==2)))*100
                    # cantidad de alumnos
                    cantidad_alumnos[n] = df_temp_new.shape[0]
                    # cantidad de cursos  = LET_CUR
                    cantidad_cursos[n] = np.unique(df_temp_new.LET_CUR).shape[0]
                    # cantidad de hombres
                    cantidad_hombres[n] = np.sum(df_temp_new.GEN_ALU==1)
                    # cantidad de mujeres
                    cantidad_mujeres[n] = np.sum(df_temp_new.GEN_ALU==2)

                else:

                    # porcentaje de hombres con respecto a mujeres
                    porcentaje_hombres[n] = np.nan
                    # cantidad de alumnos
                    cantidad_alumnos[n] = np.nan
                    # cantidad de cursos  = LET_CUR
                    cantidad_cursos[n] = np.nan
                    # cantidad de hombres
                    cantidad_hombres[n] = np.nan
                    # cantidad de mujeres
                    cantidad_mujeres[n] = np.nan


            else:
                #################################################################################################################################
                # no relizo los calculos
                promedio[n] = np.nan
                asistencia[n] = np.nan
                porcentaje_aprob[n] = np.nan
                porcentaje_nodesercion[n] = np.nan
                porcentaje_hombres[n] = np.nan
                cantidad_alumnos[n] = np.nan
                cantidad_cursos[n] = np.nan
                cantidad_hombres[n] = np.nan
                cantidad_mujeres[n] = np.nan
                pendiente_promedio[n] = np.nan

    if np.sum(id_ense==310)>0:

        # tengo que buscar cursos de media.
        for n in range(4):
            df_temp = df[(df.COD_ENSE==310) & (df.COD_GRADO==n+1)]
            if df_temp.shape[0]>0:
                #################################################################################################################################
                # realizo los calculos
                # promedio de notas
                promedio[n+8] = np.mean(df_temp.PROM_GRAL)

                # promedio asistencia
                asistencia[n+8] = np.mean(df_temp.ASISTENCIA)

                # promedio de aprobados vs reprobados
                num = (np.sum(df_temp.SIT_FIN_R=='P') - np.sum(df_temp.SIT_FIN_R=='R'))
                deno = (np.sum(df_temp.SIT_FIN_R=='P') + np.sum(df_temp.SIT_FIN_R=='R'))
                if deno>0:
                    porcentaje_aprob[n+8] = (num/deno)*100
                else:
                    porcentaje_aprob[n+8] = np.nan

                # porcentaje de retencion
                num = (np.sum(df_temp.SIT_FIN_R=='P') + np.sum(df_temp.SIT_FIN_R=='R') - np.sum(df_temp.SIT_FIN_R=='Y') - np.sum(df_temp.SIT_FIN_R=='T'))
                deno = (np.sum(df_temp.SIT_FIN_R=='P') + np.sum(df_temp.SIT_FIN_R=='R') + np.sum(df_temp.SIT_FIN_R=='Y') + np.sum(df_temp.SIT_FIN_R=='T'))
                if deno>0:
                    porcentaje_nodesercion[n+8] = (num/deno)*100
                else:
                    porcentaje_nodesercion[n+8] = np.nan

                # pendiente
                agno = np.unique(df_temp.AGNO)
                prom_vector = np.zeros(agno.shape[0])
                agno_vector = np.zeros(agno.shape[0])
                for a in range(agno.shape[0]):
                    df_temp_new = df_temp[df_temp.AGNO==agno[a]].copy()
                    prom_vector[a] = np.mean(df_temp_new.PROM_GRAL)
                    agno_vector[a] = a+1
                x = agno_vector
                y = prom_vector

                pendiente_promedio[n+8] = (x.shape[0]*(np.sum(x*y))-(np.sum(x)*np.sum(y)))/(x.shape[0]*np.sum(x**2.) - (np.sum(x))**2.)

                #################################################################################################################################
                df_temp_new = df_temp[df_temp.AGNO==2020].copy()
                if df_temp_new.shape[0]>0:

                    # porcentaje de hombres con respecto a mujeres
                    porcentaje_hombres[n+8] = (np.sum(df_temp_new.GEN_ALU==1)/(np.sum(df_temp_new.GEN_ALU==1) + np.sum(df_temp_new.GEN_ALU==2)))*100
                    # cantidad de alumnos
                    cantidad_alumnos[n+8] = df_temp_new.shape[0]
                    # cantidad de cursos  = LET_CUR
                    cantidad_cursos[n+8] = np.unique(df_temp_new.LET_CUR).shape[0]
                    # cantidad de hombres
                    cantidad_hombres[n+8] = np.sum(df_temp_new.GEN_ALU==1)
                    # cantidad de mujeres
                    cantidad_mujeres[n+8] = np.sum(df_temp_new.GEN_ALU==2)

                else:

                    # porcentaje de hombres con respecto a mujeres
                    porcentaje_hombres[n+8] = np.nan
                    # cantidad de alumnos
                    cantidad_alumnos[n+8] = np.nan
                    # cantidad de cursos  = LET_CUR
                    cantidad_cursos[n+8] = np.nan
                    # cantidad de hombres
                    cantidad_hombres[n+8] = np.nan
                    # cantidad de mujeres
                    cantidad_mujeres[n+8] = np.nan

            else:
                # no relizo los calculos
                promedio[n+8] = np.nan
                asistencia[n+8] = np.nan
                porcentaje_aprob[n+8] = np.nan
                porcentaje_nodesercion[n+8] = np.nan
                porcentaje_hombres[n+8] = np.nan
                cantidad_alumnos[n+8] = np.nan
                cantidad_cursos[n+8] = np.nan
                cantidad_hombres[n+8] = np.nan
                cantidad_mujeres[n+8] = np.nan
                pendiente_promedio[n+8] = np.nan

    elif np.sum(id_ense==410)>0:

        # tengo que buscar cursos de media.
        for n in range(4):
            df_temp = df[(df.COD_ENSE==410) & (df.COD_GRADO==n+1)]
            if df_temp.shape[0]>0:
                #################################################################################################################################
                # realizo los calculos
                # promedio de notas
                promedio[n+8] = np.mean(df_temp.PROM_GRAL)

                # promedio asistencia
                asistencia[n+8] = np.mean(df_temp.ASISTENCIA)

                # promedio de aprobados vs reprobados
                num = (np.sum(df_temp.SIT_FIN_R=='P') - np.sum(df_temp.SIT_FIN_R=='R'))
                deno = (np.sum(df_temp.SIT_FIN_R=='P') + np.sum(df_temp.SIT_FIN_R=='R'))
                if deno>0:
                    porcentaje_aprob[n+8] = (num/deno)*100
                else:
                    porcentaje_aprob[n+8] = np.nan

                # porcentaje de retencion
                num = (np.sum(df_temp.SIT_FIN_R=='P') + np.sum(df_temp.SIT_FIN_R=='R') - np.sum(df_temp.SIT_FIN_R=='Y') - np.sum(df_temp.SIT_FIN_R=='T'))
                deno = (np.sum(df_temp.SIT_FIN_R=='P') + np.sum(df_temp.SIT_FIN_R=='R') + np.sum(df_temp.SIT_FIN_R=='Y') + np.sum(df_temp.SIT_FIN_R=='T'))
                if deno>0:
                    porcentaje_nodesercion[n+8] = (num/deno)*100
                else:
                    porcentaje_nodesercion[n+8] = np.nan

                # pendiente
                agno = np.unique(df_temp.AGNO)
                prom_vector = np.zeros(agno.shape[0])
                agno_vector = np.zeros(agno.shape[0])
                for a in range(agno.shape[0]):
                    df_temp_new = df_temp[df_temp.AGNO==agno[a]].copy()
                    prom_vector[a] = np.mean(df_temp_new.PROM_GRAL)
                    agno_vector[a] = a+1
                x = agno_vector
                y = prom_vector

                pendiente_promedio[n+8] = (x.shape[0]*(np.sum(x*y))-(np.sum(x)*np.sum(y)))/(x.shape[0]*np.sum(x**2.) - (np.sum(x))**2.)

                #################################################################################################################################
                df_temp_new = df_temp[df_temp.AGNO==2020].copy()
                if df_temp_new.shape[0]>0:

                    # porcentaje de hombres con respecto a mujeres
                    porcentaje_hombres[n+8] = (np.sum(df_temp_new.GEN_ALU==1)/(np.sum(df_temp_new.GEN_ALU==1) + np.sum(df_temp_new.GEN_ALU==2)))*100
                    # cantidad de alumnos
                    cantidad_alumnos[n+8] = df_temp_new.shape[0]
                    # cantidad de cursos  = LET_CUR
                    cantidad_cursos[n+8] = np.unique(df_temp_new.LET_CUR).shape[0]
                    # cantidad de hombres
                    cantidad_hombres[n+8] = np.sum(df_temp_new.GEN_ALU==1)
                    # cantidad de mujeres
                    cantidad_mujeres[n+8] = np.sum(df_temp_new.GEN_ALU==2)

                else:

                    # porcentaje de hombres con respecto a mujeres
                    porcentaje_hombres[n+8] = np.nan
                    # cantidad de alumnos
                    cantidad_alumnos[n+8] = np.nan
                    # cantidad de cursos  = LET_CUR
                    cantidad_cursos[n+8] = np.nan
                    # cantidad de hombres
                    cantidad_hombres[n+8] = np.nan
                    # cantidad de mujeres
                    cantidad_mujeres[n+8] = np.nan

            else:
                # no relizo los calculos
                promedio[n+8] = np.nan
                asistencia[n+8] = np.nan
                porcentaje_aprob[n+8] = np.nan
                porcentaje_nodesercion[n+8] = np.nan
                porcentaje_hombres[n+8] = np.nan
                cantidad_alumnos[n+8] = np.nan
                cantidad_cursos[n+8] = np.nan
                cantidad_hombres[n+8] = np.nan
                cantidad_mujeres[n+8] = np.nan
                pendiente_promedio[n+8] = np.nan

    elif np.sum(id_ense==510)>0:

        # tengo que buscar cursos de media.
        for n in range(4):
            df_temp = df[(df.COD_ENSE==510) & (df.COD_GRADO==n+1)]
            if df_temp.shape[0]>0:
                #################################################################################################################################
                # realizo los calculos
                # promedio de notas
                promedio[n+8] = np.mean(df_temp.PROM_GRAL)

                # promedio asistencia
                asistencia[n+8] = np.mean(df_temp.ASISTENCIA)

                # promedio de aprobados vs reprobados
                num = (np.sum(df_temp.SIT_FIN_R=='P') - np.sum(df_temp.SIT_FIN_R=='R'))
                deno = (np.sum(df_temp.SIT_FIN_R=='P') + np.sum(df_temp.SIT_FIN_R=='R'))
                if deno>0:
                    porcentaje_aprob[n+8] = (num/deno)*100
                else:
                    porcentaje_aprob[n+8] = np.nan

                # porcentaje de retencion
                num = (np.sum(df_temp.SIT_FIN_R=='P') + np.sum(df_temp.SIT_FIN_R=='R') - np.sum(df_temp.SIT_FIN_R=='Y') - np.sum(df_temp.SIT_FIN_R=='T'))
                deno = (np.sum(df_temp.SIT_FIN_R=='P') + np.sum(df_temp.SIT_FIN_R=='R') + np.sum(df_temp.SIT_FIN_R=='Y') + np.sum(df_temp.SIT_FIN_R=='T'))
                if deno>0:
                    porcentaje_nodesercion[n+8] = (num/deno)*100
                else:
                    porcentaje_nodesercion[n+8] = np.nan

                # pendiente
                agno = np.unique(df_temp.AGNO)
                prom_vector = np.zeros(agno.shape[0])
                agno_vector = np.zeros(agno.shape[0])
                for a in range(agno.shape[0]):
                    df_temp_new = df_temp[df_temp.AGNO==agno[a]].copy()
                    prom_vector[a] = np.mean(df_temp_new.PROM_GRAL)
                    agno_vector[a] = a+1
                x = agno_vector
                y = prom_vector

                pendiente_promedio[n+8] = (x.shape[0]*(np.sum(x*y))-(np.sum(x)*np.sum(y)))/(x.shape[0]*np.sum(x**2.) - (np.sum(x))**2.)

                #################################################################################################################################
                df_temp_new = df_temp[df_temp.AGNO==2020].copy()
                if df_temp_new.shape[0]>0:

                    # porcentaje de hombres con respecto a mujeres
                    porcentaje_hombres[n+8] = (np.sum(df_temp_new.GEN_ALU==1)/(np.sum(df_temp_new.GEN_ALU==1) + np.sum(df_temp_new.GEN_ALU==2)))*100
                    # cantidad de alumnos
                    cantidad_alumnos[n+8] = df_temp_new.shape[0]
                    # cantidad de cursos  = LET_CUR
                    cantidad_cursos[n+8] = np.unique(df_temp_new.LET_CUR).shape[0]
                    # cantidad de hombres
                    cantidad_hombres[n+8] = np.sum(df_temp_new.GEN_ALU==1)
                    # cantidad de mujeres
                    cantidad_mujeres[n+8] = np.sum(df_temp_new.GEN_ALU==2)

                else:

                    # porcentaje de hombres con respecto a mujeres
                    porcentaje_hombres[n+8] = np.nan
                    # cantidad de alumnos
                    cantidad_alumnos[n+8] = np.nan
                    # cantidad de cursos  = LET_CUR
                    cantidad_cursos[n+8] = np.nan
                    # cantidad de hombres
                    cantidad_hombres[n+8] = np.nan
                    # cantidad de mujeres
                    cantidad_mujeres[n+8] = np.nan

            else:
                # no relizo los calculos
                promedio[n+8] = np.nan
                asistencia[n+8] = np.nan
                porcentaje_aprob[n+8] = np.nan
                porcentaje_nodesercion[n+8] = np.nan
                porcentaje_hombres[n+8] = np.nan
                cantidad_alumnos[n+8] = np.nan
                cantidad_cursos[n+8] = np.nan
                cantidad_hombres[n+8] = np.nan
                cantidad_mujeres[n+8] = np.nan
                pendiente_promedio[n+8] = np.nan

    elif np.sum(id_ense==610)>0:

        # tengo que buscar cursos de media.
        for n in range(4):
            df_temp = df[(df.COD_ENSE==610) & (df.COD_GRADO==n+1)]
            if df_temp.shape[0]>0:
                #################################################################################################################################
                # realizo los calculos
                # promedio de notas
                promedio[n+8] = np.mean(df_temp.PROM_GRAL)

                # promedio asistencia
                asistencia[n+8] = np.mean(df_temp.ASISTENCIA)

                # promedio de aprobados vs reprobados
                num = (np.sum(df_temp.SIT_FIN_R=='P') - np.sum(df_temp.SIT_FIN_R=='R'))
                deno = (np.sum(df_temp.SIT_FIN_R=='P') + np.sum(df_temp.SIT_FIN_R=='R'))
                if deno>0:
                    porcentaje_aprob[n+8] = (num/deno)*100
                else:
                    porcentaje_aprob[n+8] = np.nan

                # porcentaje de retencion
                num = (np.sum(df_temp.SIT_FIN_R=='P') + np.sum(df_temp.SIT_FIN_R=='R') - np.sum(df_temp.SIT_FIN_R=='Y') - np.sum(df_temp.SIT_FIN_R=='T'))
                deno = (np.sum(df_temp.SIT_FIN_R=='P') + np.sum(df_temp.SIT_FIN_R=='R') + np.sum(df_temp.SIT_FIN_R=='Y') + np.sum(df_temp.SIT_FIN_R=='T'))
                if deno>0:
                    porcentaje_nodesercion[n+8] = (num/deno)*100
                else:
                    porcentaje_nodesercion[n+8] = np.nan

                # pendiente
                agno = np.unique(df_temp.AGNO)
                prom_vector = np.zeros(agno.shape[0])
                agno_vector = np.zeros(agno.shape[0])
                for a in range(agno.shape[0]):
                    df_temp_new = df_temp[df_temp.AGNO==agno[a]].copy()
                    prom_vector[a] = np.mean(df_temp_new.PROM_GRAL)
                    agno_vector[a] = a+1
                x = agno_vector
                y = prom_vector

                pendiente_promedio[n+8] = (x.shape[0]*(np.sum(x*y))-(np.sum(x)*np.sum(y)))/(x.shape[0]*np.sum(x**2.) - (np.sum(x))**2.)

                #################################################################################################################################
                df_temp_new = df_temp[df_temp.AGNO==2020].copy()
                if df_temp_new.shape[0]>0:

                    # porcentaje de hombres con respecto a mujeres
                    porcentaje_hombres[n+8] = (np.sum(df_temp_new.GEN_ALU==1)/(np.sum(df_temp_new.GEN_ALU==1) + np.sum(df_temp_new.GEN_ALU==2)))*100
                    # cantidad de alumnos
                    cantidad_alumnos[n+8] = df_temp_new.shape[0]
                    # cantidad de cursos  = LET_CUR
                    cantidad_cursos[n+8] = np.unique(df_temp_new.LET_CUR).shape[0]
                    # cantidad de hombres
                    cantidad_hombres[n+8] = np.sum(df_temp_new.GEN_ALU==1)
                    # cantidad de mujeres
                    cantidad_mujeres[n+8] = np.sum(df_temp_new.GEN_ALU==2)

                else:

                    # porcentaje de hombres con respecto a mujeres
                    porcentaje_hombres[n+8] = np.nan
                    # cantidad de alumnos
                    cantidad_alumnos[n+8] = np.nan
                    # cantidad de cursos  = LET_CUR
                    cantidad_cursos[n+8] = np.nan
                    # cantidad de hombres
                    cantidad_hombres[n+8] = np.nan
                    # cantidad de mujeres
                    cantidad_mujeres[n+8] = np.nan

            else:
                # no relizo los calculos
                promedio[n+8] = np.nan
                asistencia[n+8] = np.nan
                porcentaje_aprob[n+8] = np.nan
                porcentaje_nodesercion[n+8] = np.nan
                porcentaje_hombres[n+8] = np.nan
                cantidad_alumnos[n+8] = np.nan
                cantidad_cursos[n+8] = np.nan
                cantidad_hombres[n+8] = np.nan
                cantidad_mujeres[n+8] = np.nan
                pendiente_promedio[n+8] = np.nan

    elif np.sum(id_ense==710)>0:

        # tengo que buscar cursos de media.
        for n in range(4):
            df_temp = df[(df.COD_ENSE==710) & (df.COD_GRADO==n+1)]
            if df_temp.shape[0]>0:
                #################################################################################################################################
                # realizo los calculos
                # promedio de notas
                promedio[n+8] = np.mean(df_temp.PROM_GRAL)

                # promedio asistencia
                asistencia[n+8] = np.mean(df_temp.ASISTENCIA)

                # promedio de aprobados vs reprobados
                num = (np.sum(df_temp.SIT_FIN_R=='P') - np.sum(df_temp.SIT_FIN_R=='R'))
                deno = (np.sum(df_temp.SIT_FIN_R=='P') + np.sum(df_temp.SIT_FIN_R=='R'))
                if deno>0:
                    porcentaje_aprob[n+8] = (num/deno)*100
                else:
                    porcentaje_aprob[n+8] = np.nan

                # porcentaje de retencion
                num = (np.sum(df_temp.SIT_FIN_R=='P') + np.sum(df_temp.SIT_FIN_R=='R') - np.sum(df_temp.SIT_FIN_R=='Y') - np.sum(df_temp.SIT_FIN_R=='T'))
                deno = (np.sum(df_temp.SIT_FIN_R=='P') + np.sum(df_temp.SIT_FIN_R=='R') + np.sum(df_temp.SIT_FIN_R=='Y') + np.sum(df_temp.SIT_FIN_R=='T'))
                if deno>0:
                    porcentaje_nodesercion[n+8] = (num/deno)*100
                else:
                    porcentaje_nodesercion[n+8] = np.nan

                # pendiente
                agno = np.unique(df_temp.AGNO)
                prom_vector = np.zeros(agno.shape[0])
                agno_vector = np.zeros(agno.shape[0])
                for a in range(agno.shape[0]):
                    df_temp_new = df_temp[df_temp.AGNO==agno[a]].copy()
                    prom_vector[a] = np.mean(df_temp_new.PROM_GRAL)
                    agno_vector[a] = a+1
                x = agno_vector
                y = prom_vector

                pendiente_promedio[n+8] = (x.shape[0]*(np.sum(x*y))-(np.sum(x)*np.sum(y)))/(x.shape[0]*np.sum(x**2.) - (np.sum(x))**2.)

                #################################################################################################################################
                df_temp_new = df_temp[df_temp.AGNO==2020].copy()
                if df_temp_new.shape[0]>0:

                    # porcentaje de hombres con respecto a mujeres
                    porcentaje_hombres[n+8] = (np.sum(df_temp_new.GEN_ALU==1)/(np.sum(df_temp_new.GEN_ALU==1) + np.sum(df_temp_new.GEN_ALU==2)))*100
                    # cantidad de alumnos
                    cantidad_alumnos[n+8] = df_temp_new.shape[0]
                    # cantidad de cursos  = LET_CUR
                    cantidad_cursos[n+8] = np.unique(df_temp_new.LET_CUR).shape[0]
                    # cantidad de hombres
                    cantidad_hombres[n+8] = np.sum(df_temp_new.GEN_ALU==1)
                    # cantidad de mujeres
                    cantidad_mujeres[n+8] = np.sum(df_temp_new.GEN_ALU==2)

                else:

                    # porcentaje de hombres con respecto a mujeres
                    porcentaje_hombres[n+8] = np.nan
                    # cantidad de alumnos
                    cantidad_alumnos[n+8] = np.nan
                    # cantidad de cursos  = LET_CUR
                    cantidad_cursos[n+8] = np.nan
                    # cantidad de hombres
                    cantidad_hombres[n+8] = np.nan
                    # cantidad de mujeres
                    cantidad_mujeres[n+8] = np.nan

            else:
                # no relizo los calculos
                promedio[n+8] = np.nan
                asistencia[n+8] = np.nan
                porcentaje_aprob[n+8] = np.nan
                porcentaje_nodesercion[n+8] = np.nan
                porcentaje_hombres[n+8] = np.nan
                cantidad_alumnos[n+8] = np.nan
                cantidad_cursos[n+8] = np.nan
                cantidad_hombres[n+8] = np.nan
                cantidad_mujeres[n+8] = np.nan
                pendiente_promedio[n+8] = np.nan

    elif np.sum(id_ense==810)>0:

        # tengo que buscar cursos de media.
        for n in range(4):
            df_temp = df[(df.COD_ENSE==810) & (df.COD_GRADO==n+1)]
            if df_temp.shape[0]>0:
                #################################################################################################################################
                # realizo los calculos
                # promedio de notas
                promedio[n+8] = np.mean(df_temp.PROM_GRAL)

                # promedio asistencia
                asistencia[n+8] = np.mean(df_temp.ASISTENCIA)

                # promedio de aprobados vs reprobados
                num = (np.sum(df_temp.SIT_FIN_R=='P') - np.sum(df_temp.SIT_FIN_R=='R'))
                deno = (np.sum(df_temp.SIT_FIN_R=='P') + np.sum(df_temp.SIT_FIN_R=='R'))
                if deno>0:
                    porcentaje_aprob[n+8] = (num/deno)*100
                else:
                    porcentaje_aprob[n+8] = np.nan

                # porcentaje de retencion
                num = (np.sum(df_temp.SIT_FIN_R=='P') + np.sum(df_temp.SIT_FIN_R=='R') - np.sum(df_temp.SIT_FIN_R=='Y') - np.sum(df_temp.SIT_FIN_R=='T'))
                deno = (np.sum(df_temp.SIT_FIN_R=='P') + np.sum(df_temp.SIT_FIN_R=='R') + np.sum(df_temp.SIT_FIN_R=='Y') + np.sum(df_temp.SIT_FIN_R=='T'))
                if deno>0:
                    porcentaje_nodesercion[n+8] = (num/deno)*100
                else:
                    porcentaje_nodesercion[n+8] = np.nan

                # pendiente
                agno = np.unique(df_temp.AGNO)
                prom_vector = np.zeros(agno.shape[0])
                agno_vector = np.zeros(agno.shape[0])
                for a in range(agno.shape[0]):
                    df_temp_new = df_temp[df_temp.AGNO==agno[a]].copy()
                    prom_vector[a] = np.mean(df_temp_new.PROM_GRAL)
                    agno_vector[a] = a+1
                x = agno_vector
                y = prom_vector

                pendiente_promedio[n+8] = (x.shape[0]*(np.sum(x*y))-(np.sum(x)*np.sum(y)))/(x.shape[0]*np.sum(x**2.) - (np.sum(x))**2.)

                #################################################################################################################################
                df_temp_new = df_temp[df_temp.AGNO==2020].copy()
                if df_temp_new.shape[0]>0:

                    # porcentaje de hombres con respecto a mujeres
                    porcentaje_hombres[n+8] = (np.sum(df_temp_new.GEN_ALU==1)/(np.sum(df_temp_new.GEN_ALU==1) + np.sum(df_temp_new.GEN_ALU==2)))*100
                    # cantidad de alumnos
                    cantidad_alumnos[n+8] = df_temp_new.shape[0]
                    # cantidad de cursos  = LET_CUR
                    cantidad_cursos[n+8] = np.unique(df_temp_new.LET_CUR).shape[0]
                    # cantidad de hombres
                    cantidad_hombres[n+8] = np.sum(df_temp_new.GEN_ALU==1)
                    # cantidad de mujeres
                    cantidad_mujeres[n+8] = np.sum(df_temp_new.GEN_ALU==2)

                else:

                    # porcentaje de hombres con respecto a mujeres
                    porcentaje_hombres[n+8] = np.nan
                    # cantidad de alumnos
                    cantidad_alumnos[n+8] = np.nan
                    # cantidad de cursos  = LET_CUR
                    cantidad_cursos[n+8] = np.nan
                    # cantidad de hombres
                    cantidad_hombres[n+8] = np.nan
                    # cantidad de mujeres
                    cantidad_mujeres[n+8] = np.nan

            else:
                # no relizo los calculos
                promedio[n+8] = np.nan
                asistencia[n+8] = np.nan
                porcentaje_aprob[n+8] = np.nan
                porcentaje_nodesercion[n+8] = np.nan
                porcentaje_hombres[n+8] = np.nan
                cantidad_alumnos[n+8] = np.nan
                cantidad_cursos[n+8] = np.nan
                cantidad_hombres[n+8] = np.nan
                cantidad_mujeres[n+8] = np.nan
                pendiente_promedio[n+8] = np.nan

    elif np.sum(id_ense==910)>0:

        # tengo que buscar cursos de media.
        for n in range(4):
            df_temp = df[(df.COD_ENSE==910) & (df.COD_GRADO==n+1)]
            if df_temp.shape[0]>0:
                #################################################################################################################################
                # realizo los calculos
                # promedio de notas
                promedio[n+8] = np.mean(df_temp.PROM_GRAL)

                # promedio asistencia
                asistencia[n+8] = np.mean(df_temp.ASISTENCIA)

                # promedio de aprobados vs reprobados
                num = (np.sum(df_temp.SIT_FIN_R=='P') - np.sum(df_temp.SIT_FIN_R=='R'))
                deno = (np.sum(df_temp.SIT_FIN_R=='P') + np.sum(df_temp.SIT_FIN_R=='R'))
                if deno>0:
                    porcentaje_aprob[n+8] = (num/deno)*100
                else:
                    porcentaje_aprob[n+8] = np.nan

                # porcentaje de retencion
                num = (np.sum(df_temp.SIT_FIN_R=='P') + np.sum(df_temp.SIT_FIN_R=='R') - np.sum(df_temp.SIT_FIN_R=='Y') - np.sum(df_temp.SIT_FIN_R=='T'))
                deno = (np.sum(df_temp.SIT_FIN_R=='P') + np.sum(df_temp.SIT_FIN_R=='R') + np.sum(df_temp.SIT_FIN_R=='Y') + np.sum(df_temp.SIT_FIN_R=='T'))
                if deno>0:
                    porcentaje_nodesercion[n+8] = (num/deno)*100
                else:
                    porcentaje_nodesercion[n+8] = np.nan

                # pendiente
                agno = np.unique(df_temp.AGNO)
                prom_vector = np.zeros(agno.shape[0])
                agno_vector = np.zeros(agno.shape[0])
                for a in range(agno.shape[0]):
                    df_temp_new = df_temp[df_temp.AGNO==agno[a]].copy()
                    prom_vector[a] = np.mean(df_temp_new.PROM_GRAL)
                    agno_vector[a] = a+1
                x = agno_vector
                y = prom_vector

                pendiente_promedio[n+8] = (x.shape[0]*(np.sum(x*y))-(np.sum(x)*np.sum(y)))/(x.shape[0]*np.sum(x**2.) - (np.sum(x))**2.)

                #################################################################################################################################
                df_temp_new = df_temp[df_temp.AGNO==2020].copy()
                if df_temp_new.shape[0]>0:

                    # porcentaje de hombres con respecto a mujeres
                    porcentaje_hombres[n+8] = (np.sum(df_temp_new.GEN_ALU==1)/(np.sum(df_temp_new.GEN_ALU==1) + np.sum(df_temp_new.GEN_ALU==2)))*100
                    # cantidad de alumnos
                    cantidad_alumnos[n+8] = df_temp_new.shape[0]
                    # cantidad de cursos  = LET_CUR
                    cantidad_cursos[n+8] = np.unique(df_temp_new.LET_CUR).shape[0]
                    # cantidad de hombres
                    cantidad_hombres[n+8] = np.sum(df_temp_new.GEN_ALU==1)
                    # cantidad de mujeres
                    cantidad_mujeres[n+8] = np.sum(df_temp_new.GEN_ALU==2)

                else:

                    # porcentaje de hombres con respecto a mujeres
                    porcentaje_hombres[n+8] = np.nan
                    # cantidad de alumnos
                    cantidad_alumnos[n+8] = np.nan
                    # cantidad de cursos  = LET_CUR
                    cantidad_cursos[n+8] = np.nan
                    # cantidad de hombres
                    cantidad_hombres[n+8] = np.nan
                    # cantidad de mujeres
                    cantidad_mujeres[n+8] = np.nan

            else:
                # no relizo los calculos
                promedio[n+8] = np.nan
                asistencia[n+8] = np.nan
                porcentaje_aprob[n+8] = np.nan
                porcentaje_nodesercion[n+8] = np.nan
                porcentaje_hombres[n+8] = np.nan
                cantidad_alumnos[n+8] = np.nan
                cantidad_cursos[n+8] = np.nan
                cantidad_hombres[n+8] = np.nan
                cantidad_mujeres[n+8] = np.nan
                pendiente_promedio[n+8] = np.nan


    return promedio, asistencia, porcentaje_aprob, porcentaje_nodesercion, porcentaje_hombres, cantidad_alumnos, cantidad_cursos, cantidad_hombres, cantidad_mujeres, pendiente_promedio
```
