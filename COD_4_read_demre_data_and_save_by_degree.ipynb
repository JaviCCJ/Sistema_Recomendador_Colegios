```python
def merge_data_frames(df_1,df_2, rbd):

    df_temp1 = df_1[df_1.RBD==rbd][['RBD','MRUN','COD_ENSE','COD_GRADO']]
    df_temp2 = df_2[['MRUN','PUNTAJE_PONDERADO','SIGLA_UNIVERSIDAD']]
    df_out = df_temp1[(df_temp1.COD_ENSE==310) | (df_temp1.COD_ENSE==410) | (df_temp1.COD_ENSE==510) | (df_temp1.COD_ENSE==610) | (df_temp1.COD_ENSE==710) | (df_temp1.COD_ENSE==810) | (df_temp1.COD_ENSE==910) & (df_temp1.COD_GRADO==4)].reset_index(drop=True).merge(df_temp2)
    return df_out
```
